## Task 1: dna_raw.txt

### File Handling
Reading txt file using `with statement` and one-line `for` loop

```python
# empty list to store the txt file
dna_sequences = []

# use with statement to read file named as f
with open ("dna_raw.txt", "r") as f:
    dna_sequences = [line.strip().upper() for line in f]
     # strip empty spaces and convert to upper case for every line in f
     # save cleaned data in dna_sequences
                                                                
```

```python
# store the cleaned txt file into this variable
dna_sequences
```

```python
# list for seq1, seq2, seq3, seq4
dna_id = []       

# list for A,T,C,G,N
dna_letters = []        

# for loop with startswith() function
# goes through every line in variable dna_sequence
for line in dna_sequences:
    if line.startswith('>'):              # if '>' exists argument is True
        dna_id.append(line.lstrip('>'))   # lstrip('>') will remove the string in each line given inside the argument. Append to 'dna_id' list 
                                          
        
    else:                                 # if startswith('>') argument is False
        dna_letters.append(line)          # append to 'dna_letters' list

print(f'DNA ID: {dna_id}')
print(f'DNA letters: {dna_letters}')
```


### **2. Dictionary initialization with regular expression** 
- Use regular expression `re.findall()` to find each DNA sequence (A, T, C, G) and (N).
- Count the frequency of each letter per DNA sequence. Store the dna IDs with index from 1 as dictionary keys. 
- Store the counts of each dna letters as dictionary values.


- resource regex: https://docs.python.org/3/howto/regex.html

```python
# import regex module
import re

# initialize empty dictionary for counts
letters_list = []


# enumerate() a counter as the key of enumerate object
# extract valid DNA letters from each sequence
for i, q in enumerate(dna_letters, start = 1):   # i is index, q is strings
    letters = re.findall(r"[ATCGN]", q)          # find all characters specfied inside the parameter
    letters_list.append(letters)                 # append outcome of letters

    # Print block for this sequence
    print(f"Sequence {i}:")                      # {i} is the index per  dna_id sequence, starts from 1 until 4
    print(letters)                               # remember that the result of findall(r"[ATCGN]", q) is in `letters`
    print() # for space
```

### Find frequency of character in Python
#### Count the frequency of each character by using a dictionary
- Collections library counter class: https://www.geeksforgeeks.org/python/counters-in-python-set-1/



```python
# import regular expression for pattern recognition
import re

# import collections module and Counter() function for frequency counter in dictionaries
from collections import Counter

# empty dictionary for final results
dna_numbers = {}

# temporary variable for for loop
letters_list = []

# loop through dna_sequence with enumerate for seq_number :counts
for i, q in enumerate (dna_letters, start = 1):
    letters = re.findall(r"[ATCGN]", q)    # from debbie's lecture: r for rawstring, find [ATCGN] patterns
    letters_list.append(letters)           # append the [ATCGN] patterns found with findall() to letters_list

    
    # use `letters` as parameter to count() the frequency of dna_letters 
    # convert from list to dictionary
    # and store to `counts` variable
    counts = dict(Counter(letters))

     # save in the main dictionary `dna_numbers`
     # Pattern for assignment statement in Python.
     #`variable` or `dictionary` [key] = [value] 
    dna_numbers[dna_id[i - 1]] = counts

    # mirrors the key-value statement of dictionary
    # index set to 1, counts is value
    print(dna_id[i - 1], counts)
print()
print(f'Final main dictionary: {dna_numbers}')
```

```python
import matplotlib.pyplot as plt

# loop through each sequence
for sequence_id, counts in dna_numbers.items():     #loop over both key and value
    # get letters (keys) and counts (values)
    # convert back to list for matplotlib to use it
    dna_proteins = list(counts.keys())
    dna_count = list(counts.values())

    # make a bar chart 
    plt.bar(dna_proteins, dna_count, color ='pink')

    # add labels and title
    plt.xlabel("DNA Proteins")
    plt.ylabel("DNA Counts")
    plt.title(f"Sequence ID: {sequence_id}")    # seq_id only contains SEQ1234
   
    # show each sequence's chart, iterate in for loop to show all 4
    plt.show()
```
